```python
import pandas as pd
df = pd.read_csv('/code/data/train.csv')

fake = df.loc[df['is_sarcastic'] == 1]
real = df.loc[df['is_sarcastic'] == 0]
```

```python
fake_len = fake['headline'].astype(str).map(len)
real_len = real['headline'].astype(str).map(len)
```

```python
import numpy as np; np.random.seed(0)
import matplotlib.pyplot as plt
import scipy.stats

a = fake_len
b = real_len

kdea = scipy.stats.gaussian_kde(a)
kdeb = scipy.stats.gaussian_kde(b)

grid = np.linspace(0,150)

plt.plot(grid, kdea(grid), label="Fake news length")
plt.plot(grid, kdeb(grid), label="Real news length")

plt.legend()
plt.show()
```
